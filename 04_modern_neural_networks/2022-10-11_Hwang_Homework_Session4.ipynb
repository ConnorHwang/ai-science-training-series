```python
%matplotlib inline

import matplotlib.pyplot as plt
import os

# File location
pathname = os.path.abspath('.')
filename = 'output.dat'

# Arrays
index = []
epoch = []
loss = []
acc = []

# Read file
try:
  fileRead = open(os.path.join(pathname,filename), 'r')
except:
  print('WARNING - File not present: ' + os.path.join(pathname,filename))
else:
  data = fileRead.read()
  fileRead.close()
  data = data.split('\n')

for i in range(len(data)-1):
  line = data[i].split()
#   line = line.split()
#   line = re.split(r'\t+', line)
#   line = line.split('\t')
  

  # append
  index.append(float(line[0]))
  epoch.append(float(line[1]))
  loss.append(float(line[2]))
  acc.append(float(line[3]))

# plotting
pltsize = 5
plt.figure(figsize=(2*pltsize, pltsize))
plt.subplot(1,2,1)
plt.plot(range(len(loss)), loss)
plt.grid()
plt.title('Loss')

plt.subplot(1,2,2)
plt.plot(range(len(acc)), acc)
plt.grid()
plt.title('Accuracy')
```
